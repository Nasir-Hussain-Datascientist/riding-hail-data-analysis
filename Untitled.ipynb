```python
df_cleaned = df.dropna(subset=['metered_price', 'upfront_price', 'distance', 'duration', 'predicted_distance', 'predicted_duration', 'gps_confidence', 'dest_change_number'])
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import matplotlib.pyplot as plt

# Assume df_cleaned is your DataFrame

# Fill null values in 'fraude_source' with 0
df_cleaned['fraud_score'] = df_cleaned['fraud_score'].fillna(0)

# Prepare the data
X = df_cleaned[['metered_price', 'upfront_price', 'distance', 'duration', 'predicted_distance', 'predicted_duration', 'gps_confidence', 'dest_change_number', 'fraud_score']]
y = df_cleaned['overpaid_ride_ticket']

# Initialize and fit the model
model = RandomForestClassifier()
model.fit(X, y)

# Get feature importances
importance = model.feature_importances_
feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': importance})
print(feature_importances.sort_values(by='Importance', ascending=False))

# Sort the features by importance
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importances)
plt.title('Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
import statsmodels.api as sm

X = df_cleaned[['metered_price', 'upfront_price', 'distance', 'duration', 'predicted_distance', 'predicted_duration', 'gps_confidence', 'dest_change_number']]
y = df_cleaned['overpaid_ride_ticket']

X = sm.add_constant(X)  # Adds a constant term to the predictor
model = sm.Logit(y, X)
result = model.fit()
print(result.summary())
# Extract coefficients
coefficients = result.params
coefficients_df = pd.DataFrame({'Feature': coefficients.index, 'Coefficient': coefficients.values})

# Sort the coefficients by their absolute value for better visualization
coefficients_df['Absolute Coefficient'] = coefficients_df['Coefficient'].abs()
coefficients_df = coefficients_df.sort_values(by='Absolute Coefficient', ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Coefficient', y='Feature', data=coefficients_df)
plt.title('Feature Coefficients')
plt.xlabel('Coefficient')
plt.ylabel('Feature')
plt.show()
# Predict probabilities
df_cleaned['predicted_prob'] = result.predict(X)

# Plot the distribution of predicted probabilities
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['predicted_prob'], bins=30, kde=True)
plt.title('Distribution of Predicted Probabilities')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.show()
# Plot actual vs predicted
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_cleaned['predicted_prob'], y=df_cleaned['overpaid_ride_ticket'])
plt.title('Actual vs Predicted Probabilities')
plt.xlabel('Predicted Probability')
plt.ylabel('Actual Value')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the data
X = df_cleaned[['metered_price', 'upfront_price', 'distance', 'duration', 'predicted_distance', 'predicted_duration']]

# Initialize and fit the model
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)  # Set n_init explicitly
df_cleaned['cluster'] = kmeans.fit_predict(X)

# Plot the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='metered_price', y='upfront_price', hue='cluster', palette='viridis')
plt.title('Clusters based on KMeans')
plt.xlabel('Metered Price')
plt.ylabel('Upfront Price')
plt.legend(title='Cluster')
plt.show()
```

```python
# Group by cluster and compute descriptive statistics
cluster_summary = df_cleaned.groupby('cluster').agg({
    'metered_price': ['mean', 'std', 'min', 'max'],
    'upfront_price': ['mean', 'std', 'min', 'max'],
    'distance': ['mean', 'std', 'min', 'max'],
    'duration': ['mean', 'std', 'min', 'max'],
    'predicted_distance': ['mean', 'std', 'min', 'max'],
    'predicted_duration': ['mean', 'std', 'min', 'max']
}).reset_index()

# Rename columns for easier understanding
cluster_summary.columns = ['Cluster', 'Metered Price Mean', 'Metered Price Std', 'Metered Price Min', 'Metered Price Max',
                           'Upfront Price Mean', 'Upfront Price Std', 'Upfront Price Min', 'Upfront Price Max',
                           'Distance Mean', 'Distance Std', 'Distance Min', 'Distance Max',
                           'Duration Mean', 'Duration Std', 'Duration Min', 'Duration Max',
                           'Predicted Distance Mean', 'Predicted Distance Std', 'Predicted Distance Min', 'Predicted Distance Max',
                           'Predicted Duration Mean', 'Predicted Duration Std', 'Predicted Duration Min', 'Predicted Duration Max']

print(cluster_summary)
```
